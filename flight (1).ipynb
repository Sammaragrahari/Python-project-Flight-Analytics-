#Q1.What are the airlines in the dataset, accompanied by the frequencies?

```python
# Showing all the Aiorlines with their Number of flight in bar graph

data["airline"].value_counts(ascending=True).plot.barh( color= ["lightgreen","lightblue"])

plt.title("Airlines with frequencies")

plt.xlabel("Number of flight")

plt.ylabel("Airlines")

plt.show()
```

#Q2. Show the bar graphs representing the Daparture time & Arrival time


```python
data.head()
```

```python
# Showing the Departure time for the flights
data["departure_time"].value_counts()
```

```python
# Showing the Arrival time for the flight

data["arrival_time"].value_counts()
```

```python
# Showing the Depaarture Time & Arrival Time for the flights with their counts

plt.figure(figsize=(16,4))   

# --- Departure Time ---
plt.subplot(1,2,1)
dep = data["departure_time"].value_counts()

plt.bar(dep.index, dep.values, color=['g','y'])
plt.xlabel("Departure Time")
plt.ylabel("D.Freq")
plt.title("Departure Time Frequency")
plt.xticks(rotation=45)

# --- Arrival Time ---
plt.subplot(1,2,2)
arr = data["arrival_time"].value_counts()

plt.bar(arr.index, arr.values, color=['r','b'])
plt.xlabel("Arrival Time")
plt.ylabel("D.Freq")
plt.title("Arrival Time Frequency")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
